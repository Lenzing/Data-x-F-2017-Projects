```python
# stretch Jupyter coding blocks to fit screen
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:75% !important; }</style>")) # if 100% it would fit the screen
# make it run on py2 and py3
from __future__ import division, print_function
import requests # The requests library is an 
# HTTP library for getting content and posting etc.

import bs4 as bs # BeautifulSoup4 is a Python library 
# for pulling data out of HTML and XML code.
import pandas as pd
```

```python
source = requests.get("https://blockchain.info/charts").content 
# get the source content
soup = bs.BeautifulSoup(source,'lxml')



#https://api.blockchain.info/charts/n-transactions?format=csv
links = soup.find_all("a")
chartlinks=[]
chartname=[]

#get all of the links to the csv files as well as the chart names 
for l in links:
    this = l.get("href")
    if "/charts/" in this:
        chartlinks.append("https://blockchain.info"+this+"?format=csv")
        chartname.append(this[8:])
#pd.read_csv(chartlinks[0],header=None).head()
names =[]
links =[]

#remove any duplicates that may be present
for this in chartname:
    if this not in names:
        names.append(this)
len(names)
for this in chartlinks:
    if this not in links:
        links.append(this)
chartname=names
chartlinks=links
#print(names,"\n",links)

data=pd.read_csv(chartlinks[0],header = None,)
data=data.rename(columns={0:"Date",1:chartname[0]})
data.index = pd.to_datetime(data["Date"])
data.drop(["Date"],axis=1,inplace=True)

#create a combined dataframe with features from each blockchain chart 
for i,link in enumerate(chartlinks[1:]):
    try:
        #print(i,": ",chartname[i+1])    
        b=pd.read_csv(link,header = None,)
        b=b.rename(columns={0:"Date",1:chartname[i+1]})
        b.index = pd.to_datetime(b["Date"])
        b.drop(["Date"],axis=1,inplace=True)
        #print(a.head())
        #print(b.head())
        if b.size == 365:
            data=data.join(b,how="outer")
        
    except:
        print(chartname[i+1], " could not be found")
    
#data.head()

    
    

```
